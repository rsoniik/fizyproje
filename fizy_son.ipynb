```python
import pandas as pd

## örnek bir veri seti oluşturma (kullanıcıların örnek listeleri)

kullanici1 = pd.read_csv("kullanici1.txt") 
kullanici2 = pd.read_csv("kullanici2.txt") ## Ayşe
kullanici3 = pd.read_csv("kullanici3.txt") ## Osman
kullanici4 = pd.read_csv("kullanici4.txt") ## Mehmet
kullanici5 = pd.read_csv("kullanici5.txt") ## Resul

kullanici1df = pd.DataFrame(kullanici1)
kullanici2df = pd.DataFrame(kullanici2)
kullanici3df = pd.DataFrame(kullanici3)
kullanici4df = pd.DataFrame(kullanici4)
kullanici5df = pd.DataFrame(kullanici5)


song_list1 = kullanici1df['Tur'].tolist()
song_list2 = kullanici2df['Tur'].tolist()
song_list3 = kullanici3df['Tur'].tolist()
song_list4 = kullanici4df['Tur'].tolist()
song_list5 = kullanici5df['Tur'].tolist()




def calculate_genre_ratio(user):
    
    tr_pop = user.count('Turkish Pop')
    dc_el = user.count('Dans/Elektronik')
    hh_rap = user.count('Hip-Hop/Rap')
    tr_rck = user.count('Turkish Rock')
    alt = user.count('Alternative')
    pop = user.count('Pop')
    el_swg = user.count('Elektro Swing')
    i_rck = user.count('Indie Rock')
    arbsk = user.count('Arabesk')
    pop_rck = user.count('Pop Rock')
    el_swg = user.count('Elektro Swing')
    metal = user.count('Metal')
    hrd_rck = user.count('Hard Rock')
    rck = user.count('Rock')
    tr_flk = user.count('Turkish Folk')

    pop_rt = tr_pop + pop + pop_rck
    elc_rt = dc_el + el_swg
    rap_rt = hh_rap 
    rck_rt = tr_rck + i_rck + hrd_rck + rck + pop_rck
    alt_rt = alt
    arbsk_rt = arbsk
    metal_rt = metal
    trflk_rt = tr_flk

    return [pop_rt, elc_rt, rap_rt, rck_rt, alt_rt, arbsk_rt, metal_rt, trflk_rt]
   
def test_compatibility(user1, user2):
    
    list1 = calculate_genre_ratio(user1)
    list2 = calculate_genre_ratio(user2) 

    k = 1
    n = 0
        
    while k<8:
        m1 = list1[n]
        m2 = list2[n]
        max1 = max(list1)
        max2 = max(list2)

        if m1 == m2:
            
            if max1 == 0:
                k = k+1
                n = n+1

            elif m1 == max1:
                if n == 0:
                    a = "İkinizin de en sevdiği tür pop!"
                    return a

                elif n == 1:
                    b = "İkinizin de en sevdiği tür elektronik!"
                    return b

                elif n == 2:
                    c = "İkinizin de en sevdiği tür rap!"
                    return c

                elif n == 3:
                    d = "İkinizin de en sevdiği tür rock!"
                    return d

                elif n == 4:
                    e = "İkinizin de en sevdiği tür alternatif!"
                    return e

                elif n == 5:
                    f = "İkinizin de en sevdiği tür arabesk!"
                    return f

                elif n == 6:
                    g = "İkinizin de en sevdiği tür metal!"
                    return g

                elif n == 7:
                    h = "İkinizin de en sevdiği tür Türk halk müziği!"
                    return h
            else:
                i = "Pek de ortak bir müzik zevkiniz yok."
                return i
        
        elif m1 > m2:
            if abs(m1 - m2) == 1 or abs(m1 - m2) == 2 and m1 == max1:
                if n == 0:
                    a = "İkinizin de en sevdiği tür pop!"
                    return a

                elif n == 1:
                    b = "İkinizin de en sevdiği tür elektronik!"
                    return b

                elif n == 2:
                    c = "İkinizin de en sevdiği tür rap!"
                    return c

                elif n == 3:
                    d = "İkinizin de en sevdiği tür rock!"
                    return d

                elif n == 4:
                    e = "İkinizin de en sevdiği tür alternatif!"
                    return e

                elif n == 5:
                    f = "İkinizin de en sevdiği tür arabesk!"
                    return f

                elif n == 6:
                    g = "İkinizin de en sevdiği tür metal!"
                    return g

                elif n == 7:
                    h = "İkinizin de en sevdiği tür Türk halk müziği!"
                    return h
                
                else:
                    i = "Pek de ortak bir müzik zevkiniz yok."
                    return i
            
            elif abs(m1 - m2) > 2:
                i = "Pek de ortak bir müzik zevkiniz yok."
                return i
            
        else:
            i = "Pek de ortak bir müzik zevkiniz yok."
            return i

## sonuç        
x = input(str("""Hangi kullanıcıyla müzik zevkinizi karşılaştırmak istersiniz? 
            Bulunan Kullanıcılar:
            Ayşe
            Osman
            Mehmet 
            Resul\n"""))
          
if x == "Ayşe":
    print(test_compatibility(song_list1, song_list2))

elif x == "Osman":
     print(test_compatibility(song_list1, song_list3))

elif x == "Mehmet":
    print(test_compatibility(song_list1, song_list4))

elif x == "Resul":
    print(test_compatibility(song_list1, song_list5))

else:
    print("Yanlış isim girdiniz. Lütfen tekrar deneyiniz.")
```
